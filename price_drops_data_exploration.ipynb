```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from pandas.api.types import is_numeric_dtype
```

```python
price_drops_csv = "redfin_price_drops_weekly_all_country_2021_Jan_to_2026_May.csv"
df = pd.read_csv(price_drops_csv)
df["LAST UPATED"] = pd.to_datetime(df['LAST UPDATED'])
df["PERIOD BEGIN"] = pd.to_datetime(df['PERIOD BEGIN'])
df["PERIOD END"] = pd.to_datetime(df['PERIOD END'])
df['LAST UPATED'] = pd.to_datetime(df['LAST UPATED'])
df = df.sort_values("PERIOD BEGIN").reset_index(drop=True)
```

```python
df.head()
```

```python
for col in df.columns:
    if not is_numeric_dtype(df[col]):
        continue
    sns.scatterplot(data=df, x='PERIOD BEGIN', y=col)
    plt.xticks(rotation=45, ha='right')
    plt.show()
```
